# Packages

```python

from matplotlib import pyplot as plt
from skimage import data

# Transformations
from skimage import transform
import numpy as np
```

```python
def plot_comparison(original, sobel_image, title):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6))
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title("Original")
    ax2.imshow(sobel_image, cmap=plt.cm.gray)
    ax2.set_title(title)
    plt.show()
```

```python
# Get Image from Skimage Data
img = data.cat()
plt.imshow(img, cmap=plt.cm.gray)
plt.title(f'Shape: {img.shape}')
```

# Rotations

```python
# Apply ClockWise Rotation
rotated_img = transform.rotate(img, -90)
plot_comparison(img, rotated_img, "Clockwise Rotation")
```

# Rescale

```python
# Apply Rescale 1/4

rescaled_img = transform.rescale(img, 1/4, anti_aliasing=True, channel_axis=2)
plot_comparison(img, rescaled_img, "Rescaled Image")
```
