# Packages

```python
from skimage import data
import numpy as np
from matplotlib import pyplot as plt
```

```python
def show_image(img, title = "", cmap = "gray"):
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# Get Sample Image
my_image = data.camera()
print(my_image.shape)
show_image(my_image, title="Original")
```

```python
thresh = 127

binary = my_image > thresh

show_image(my_image, "Original")
show_image(binary, "Thresholded")
```

# Functions for Threshold

```python
# Global Threshold
from skimage.filters import threshold_otsu
```

```python
thresh_otsu = threshold_otsu(my_image)
binary_with_otsu = my_image > thresh_otsu
show_image(binary_with_otsu, "Threshold Otsu")
```

```python
# Local Threshold
from skimage.filters import threshold_local
text_img =  data.page()
show_image(text_img, "original")
```

```python
text_thresh_value = threshold_otsu(text_img)
show_image(text_img>text_thresh_value, "otsu")
```

```python
# Set Region Pixel
block_size = 35

# Get local Thresh
local_thresh = threshold_local(text_img, block_size, offset=10)

# Apply local Thresh
binary_result = text_img > local_thresh
show_image(binary_result, "Local Thresh")
```
