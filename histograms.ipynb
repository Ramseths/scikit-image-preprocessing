# Packages

```python
from skimage import data
import numpy as np
from matplotlib import pyplot as plt
```

# Get Image

```python
my_image = data.rocket()
plt.imshow(my_image)
```

```python
# Convert to Grayscale (N, N)
from skimage import color
my_image_grayscale = color.rgb2gray(my_image)
plt.imshow(my_image_grayscale)
```

# Histogram

```python
red = my_image[:,:,0]
plt.hist(red.ravel(), bins=256, color='red')
plt.show()
```

```python
green = my_image[:,:,1]
plt.hist(green.ravel(), bins=256, color='green')
plt.show()
```

```python
blue = my_image[:,:,1]
plt.hist(blue.ravel(), bins=256, color='blue')
plt.show()
```
