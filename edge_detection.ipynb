# Packages

```python
from matplotlib import pyplot as plt
from skimage import color
from skimage import data

# Filter Sobel
from skimage.filters import sobel
import numpy as np
```

```python
# Get Image
coins = data.coins()

# Show image
plt.imshow(coins, cmap="gray")
plt.axis("off")
plt.show()
```

```python
def plot_comparison(original, sobel_image, title):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True, sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title("Original")
    ax1.axis("off")
    ax2.imshow(sobel_image, cmap=plt.cm.gray)
    ax2.set_title("Sobel")
    ax2.axis("off")
    plt.show()
```

```python
# Apply Sobel Function
edge_sobel = sobel(coins)

plot_comparison(coins, edge_sobel, "Sobel Function")
```

# Gaussian Filter for Smoothing

```python
# Get Filter
from skimage.filters import gaussian

# Get Image
img = data.astronaut()

# Apply Smoothing
img_smoothing = gaussian(img, channel_axis=2)

plot_comparison(img, img_smoothing, "Smoothing Image")
```
